```python
import os

def get_file_paths_and_labels(base_dir):
    file_paths = []
    labels = []
    
    # Define the folder names and their corresponding labels
    datasets = [
        ('negative_dataset_1sec', 0),
        ('negative_dataset_10sec', 0),
        ('positive_dataset_1sec', 1),
        ('positive_dataset_10sec', 1)
    ]
    
    for dataset, label in datasets:
        dataset_dir = os.path.join(base_dir, dataset)
        print(f"Checking directory: {dataset_dir}")  # Debugging print
        for root, _, files in os.walk(dataset_dir):
            print(f"Currently in root: {root}")  # Debugging print
            for file in files:
                file_path = os.path.join(root, file)
                # Check if the file is an MP3 file by its type
                if os.path.isfile(file_path):
                    print(f"Found file: {file_path}")  # Debugging print
                    file_paths.append(file_path)
                    labels.append(label)
    
    return file_paths, labels

# base_dir = r'/home/gridsan/abradshaw/MIT Buzz'  # Update with the actual path
# file_paths, labels = get_file_paths_and_labels(base_dir)

# print(f"Found {len(file_paths)} audio files.")
```

```python
# import librosa
# import numpy as np
# import torch
# from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

# def load_and_resample_audio(file_path, target_sr=16000):
#     try:
#         print(f"Loading audio file: {file_path}")  # Debugging print
#         y, sr = librosa.load(file_path, sr=None, mono=False)
#         print(f"Original sample rate: {sr}, Target sample rate: {target_sr}")  # Debugging print
#         if y.ndim > 1:
#             y = librosa.to_mono(y)  # Convert to mono if stereo
#             print("Converted to mono")  # Debugging print
#         y_resampled = librosa.resample(y, orig_sr=sr, target_sr=target_sr)
#         print(f"Resampled audio to {target_sr} Hz")  # Debugging print
#         return y_resampled, target_sr
#     except Exception as e:
#         print(f"Error loading or resampling audio: {e}")
#         return None, None

# def extract_mfcc(y, sr, n_mfcc=13, n_fft=2048, hop_length=512):
#     try:
#         print(f"Extracting MFCC features")  # Debugging print
#         mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length)
#         mfcc = (mfcc - np.mean(mfcc)) / np.std(mfcc)  # Standardize the MFCCs
#         print(f"Extracted MFCC shape: {mfcc.shape}")  # Debugging print
#         return mfcc.T  # Transpose to get (time, n_mfcc) shape
#     except Exception as e:
#         print(f"Error extracting MFCC: {e}")
#         return None

# # Update with the actual path
# base_dir = r'/home/gridsan/abradshaw/MIT Buzz'
# file_paths, labels = get_file_paths_and_labels(base_dir)

# print(f"Found {len(file_paths)} audio files.")

# # Process each file
# for file_path, label in zip(file_paths, labels):
#     y, sr = load_and_resample_audio(file_path)
#     if y is not None:
#         mfcc_features = extract_mfcc(y, sr)
#         if mfcc_features is not None:
#             print(f"MFCC Features for {file_path}: {mfcc_features.shape}")
#         else:
#             print(f"MFCC extraction failed for {file_path}.")
#     else:
#         print(f"Audio loading failed for {file_path}.")

import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

def load_and_resample_audio(file_path, target_sr=48000):
    try:
        print(f"Loading audio file: {file_path}")  # Debugging print
        y, sr = librosa.load(file_path, sr=None, mono=False)
        print(f"Original sample rate: {sr}, Target sample rate: {target_sr}")  # Debugging print
        if y.ndim > 1:
            y = librosa.to_mono(y)  # Convert to mono if stereo
            print("Converted to mono")  # Debugging print
        y_resampled = librosa.resample(y, orig_sr=sr, target_sr=target_sr)
        print(f"Resampled audio to {target_sr} Hz")  # Debugging print
        return y_resampled, target_sr
    except Exception as e:
        print(f"Error loading or resampling audio: {e}")
        return None, None

def extract_mfcc(y, sr, n_mfcc=13, n_fft=2048, hop_length=512):
    try:
        print(f"Extracting MFCC features")  # Debugging print
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length)
        mfcc = (mfcc - np.mean(mfcc)) / np.std(mfcc)  # Standardize the MFCCs
        print(f"Extracted MFCC shape: {mfcc.shape}")  # Debugging print
        return mfcc.T  # Transpose to get (time, n_mfcc) shape
    except Exception as e:
        print(f"Error extracting MFCC: {e}")
        return None

def visualize_spectrogram(y, sr, title="Spectrogram"):
    plt.figure(figsize=(10, 4))
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)
    S_dB = librosa.power_to_db(S, ref=np.max)
    librosa.display.specshow(S_dB, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()

def visualize_mfcc(mfcc, sr, hop_length, title="MFCC"):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfcc, sr=sr, hop_length=hop_length, x_axis='time')
    plt.colorbar()
    plt.title(title)
    plt.tight_layout()
    plt.show()

# Assuming get_file_paths_and_labels is defined somewhere else in your script
base_dir = r'/home/gridsan/abradshaw/MIT Buzz/'
file_paths, labels = get_file_paths_and_labels(base_dir)

print(f"Found {len(file_paths)} audio files.")

# Process only the first five files
for idx, (file_path, label) in enumerate(zip(file_paths[:5], labels[:5])):
    y, sr = load_and_resample_audio(file_path)
    if y is not None:
        mfcc_features = extract_mfcc(y, sr)
        if mfcc_features is not None:
            print(f"MFCC Features for {file_path}: {mfcc_features.shape}")
            visualize_spectrogram(y, sr, title=f"Spectrogram for {file_path}")
            visualize_mfcc(mfcc_features.T, sr, hop_length=512, title=f"MFCC for {file_path}")
        else:
            print(f"MFCC extraction failed for {file_path}.")
    else:
        print(f"Audio loading failed for {file_path}.")


```

```python
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

class MFCCDataset(Dataset):
    def __init__(self, file_paths, labels, n_mfcc=13):
        self.file_paths = file_paths
        self.labels = labels
        self.n_mfcc = n_mfcc

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        file_path = self.file_paths[idx]
        label = self.labels[idx]
        y, sr = load_and_resample_audio(file_path)
        mfcc_features = extract_mfcc(y, sr, n_mfcc=self.n_mfcc)
        mfcc_features = torch.tensor(mfcc_features, dtype=torch.float32)
        label_tensor = torch.tensor(label, dtype=torch.float32)
        return mfcc_features, label_tensor

# Create the dataset
dataset = MFCCDataset(file_paths, labels)

# Use WeightedRandomSampler if needed to handle class imbalance
weights = [5.0 if label == 1 else 1.0 for label in labels]
sampler = WeightedRandomSampler(weights, len(weights), replacement=True)

# Create DataLoader
train_loader = DataLoader(dataset, batch_size=32, sampler=sampler)
test_loader = DataLoader(dataset, batch_size=32, shuffle=False)
```

```python
from pytorch_tcn import TCN
import torch.nn as nn
import torch.optim as optim

# Define the TCN model
model = TCN(
    num_inputs=13,
    num_channels=[25, 50, 50, 100],
    kernel_size=7,
    dilations=None,
    dropout=0.3,
    causal=False,
    use_norm='weight_norm',
    activation='relu',
    kernel_initializer='xavier_uniform',
    use_skip_connections=False,
    input_shape='NCL',
    embedding_shapes=None,
    embedding_mode='add',
    use_gate=False,
    lookahead=0,
    output_projection=None,
    output_activation=None
)

# Wrap the model with DataParallel if using multiple GPUs
if torch.cuda.device_count() > 1:
    model = nn.DataParallel(model)

# Move the model to the appropriate device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)

# Training loop
num_epochs = 20

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs.squeeze(), y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    epoch_loss /= len(train_loader)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
```
